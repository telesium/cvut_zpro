# Vývojové prostředí (IDE)

Dosud jsme k programování využívali prostředí Jupyter v prohlížeči – platforma JupyterHub.
Dnes se však naučíme používat plnohodnotné _vývojové prostředí_ (IDE), které je v podstatě základním nástrojem programátora.
Vývojové prostředí je samostatný program spouštěný v operačním systému, který je základním rozhraním pro komunikaci mezi uživatelem a počítačem.

Abychom si zjednodušili komunikaci mezi sebou, budeme všichni používat stejné vývojové prostředí, které do jisté míry odstíní rozdíly mezi různými operačními systémy.
Budeme používat **VSCode**, resp. **VSCodium**.

## VSCode a VSCodium

[VSCode](https://code.visualstudio.com/) a [VSCodium](https://vscodium.com/) jsou vývojová prostředí, která vychází ze společného základu a mají tedy většinu stejných vlastností: podporují všechny tři hlavní operační systémy, umožňují programovat v podstatě v libovolném programovacím jazyku, lze je modifikovat pomocí rozšíření (*extensions*), atd.

Z pohledu uživatelů (programátorů) jsou zásadní tyto rozdíly mezi VSCode a VSCodium:

1. VSCode je uzavřený produkt Microsoftu, který je distribuovaný v binární formě **se zapnutou telemetrií** (sledování uživatelů), kdežto VSCodium je svobodný komunitní projekt, kde je telemetrie vypnuta.
2. Prostředí VSCode a VSCodium používají různé _marketplace_ pro rozšíření: [VSCode Marketplace](https://marketplace.visualstudio.com/VSCode) od Microsoftu vs [Open VSX Registry](https://open-vsx.org/), který funguje i pro svobodné varianty (VSCodium).

Na školních počítačích v učebnách je nainstalovaný VSCode (v systémové složce `C:\Program Files\`).
My si však nainstalujeme **VSCodium**:

- instalace pro Windows: viz dále
- instalace pro macOS: [pomocí příkazu brew](https://vscodium.com/#install-with-brew-mac)
  nebo [release pro x64](https://github.com/VSCodium/vscodium/releases/download/1.105.16954/VSCodium.x64.1.105.16954.dmg)
  nebo [release pro arm64](https://github.com/VSCodium/vscodium/releases/download/1.105.16954/VSCodium.arm64.1.105.16954.dmg)
- instalace pro Linux: buď jako [systémový balíček](https://vscodium.com/#use-a-package-manager-deb-rpm-provided-by-vscodium-related-repos)
  nebo stáhnout [release z GitHubu](https://github.com/VSCodium/vscodium/releases)
  (např. [verze 1.105.16954](https://github.com/VSCodium/vscodium/releases/download/1.105.16954/VSCodium-linux-x64-1.105.16954.tar.gz))

# Želví grafika

Želví grafika je populární nástroj pro kreslení obrázků pomocí programu: programátor ovládá želvu, která se plazí po pláži a v písku za sebou nechává stopu.
Tento koncept byl navržen pro výuku v programovacím jazyku Logo.

Želví grafika je v Pythonu dostupná v modulu `turtle`.
Základní funkce pro ovládání želvy jsou:

- `forward(distance)` – posun dopředu o zadanou vzdálenost (v pixelech)
- `backward(distance)` – posun dozadu o zadanou vzdálenost (v pixelech)
- `left(angle)` – otočení doleva o zadaný úhel (defaultně ve stupních)
- `right(angle)` – otočení doprava o zadaný úhel (defaultně ve stupních)

Funkce pro další akce, jako např. změnu barvy při kreslení, najdete v [dokumentaci k modulu `turtle`](https://docs.python.org/3/library/turtle.html).

## Příklad

Vytvořte program s následujícím kódem:

```python
import turtle

# draw a square
for i in range(4):
    turtle.forward(100)
    turtle.right(90)

# wait until the user closes the window
turtle.mainloop()
```

**Pozor:** Váš program se nesmí jmenovat `turtle.py` – jinak by překryl standardní knihovnu a příkaz `import` by importoval váš program místo standardního modulu!

**Poznámka:** Pokud program skončí kvůli chybě `ModuleNotFoundError: No module named 'tkinter'`, je potřeba ještě doinstalovat modul `tkinter` pro grafické rozhraní.
Při instalaci ve Windows (viz výše) je modul `tkinter` automaticky k dispozici, v macOS se dá doinstalovat příkazem `brew install python-tk`, pro Linuxové distribuce existuje mnoho variant – viz [StackOverflow](https://stackoverflow.com/q/76105218).

Program můžete spustit buď z vývojového prostředí, což je výhodné při programování, nebo přímo z operačního systému, což je výhodné pro hotové programy:
